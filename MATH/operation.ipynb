# Broadcast


```python
import numpy as np

A = np.array([[1, 3, 5],
[2, 4, 6]])
v = np.array([[1], [2]])

# Now you can add
result = A + v

print(result)
```

# Multiplication


```python
a = np.array([[1,2,3], [4,5,6]])
b = np.array([[1,2], [3,1], [2,3]])

result = a.dot(b)
print(result)
```

# Transpose
